### Getting the most recent data

```python
indicator_name = 'GDP Growth Rate'
df_investing_indicator = data_provider.get_indicator_values(indicator_name, 'Investing', 'QS')
df_oecd_indicator = data_provider.get_indicator_values(indicator_name, 'OECD', 'QS')
df_gdp_combined = df_investing_indicator.combine_first(df_oecd_indicator)
df_gdp_combined
```

```python
date = pd.Timestamp('2023-12-06')
date_minus_2_quarters = date - pd.DateOffset(months=6)

latest_known_period = pd.to_datetime(f'{date_minus_2_quarters.year}-Q{date_minus_2_quarters.quarter}')
df_latest = df_gdp_combined[:latest_known_period].iloc[-int(6/3):]
df_latest
```

```python
new_monthly_index = pd.date_range(df_latest.index[0], df_latest.index[-1]+pd.DateOffset(months=2), freq='MS')
df_mon = pd.DataFrame(df_latest, index=new_monthly_index)
df_mon.ffill()
```

## Fill missing values

I believe the most reliable way would be to get the list of correlated countries and fill missing values with mean values of 5 most correlated countries.

```python
indicator_name = 'GDP Growth Rate'
df_investing_indicator = data_provider.get_indicator_values(indicator_name, 'Investing', 'QS')
df_oecd_indicator = data_provider.get_indicator_values(indicator_name, 'OECD', 'QS')
df = df_investing_indicator.combine_first(df_oecd_indicator)
df
```

```python
corr = data_provider.calculate_correlations_for_returns()
corr
```

```python
df = df.ffill(limit=3).bfill(limit=3)
countries_with_missing_data = df.columns[df.isna().sum() > 0]
countries_with_missing_data
```

```python
country = countries_with_missing_data[0]
most_correlated_countries = corr[country].sort_values()[::-1]
most_correlated_countries
```

```python
most_correlated_countries = corr[country].sort_values()[::-1][1:].index
most_correlated_countries_with_data = most_correlated_countries[~most_correlated_countries.isin(countries_with_missing_data)][:5]
most_correlated_countries_with_data
```

```python
missing_dates = df[df[country].isna()].index
mean_values = df.loc[missing_dates, most_correlated_countries_with_data].mean(axis=1).round(2)
mean_values
```

```python
df.loc[missing_dates, country] = mean_values
```
