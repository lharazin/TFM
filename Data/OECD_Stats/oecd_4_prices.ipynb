# OECD Indicators - Prices

Exploring all indicators for Prices category from OECD. Described in chapter 3.2.

```python
import os
os.chdir(os.path.expanduser('../AWS_Lambdas/src/lambda_oecd_download/'))
from oecd_api import get_oecd_data
```

```python
indicator = 'KEI' # Key Short-Term Economic Indicators
subject = 'CPALTT01' # Consumer Prices - Annual inflation
measure = 'GY' # Growth on the same period of the previous year
freq = 'M' # Monthly
df = get_oecd_data(indicator, subject, measure, freq, 
                   f'../../../OECD_Stats/Indicators/OECD_Inflation Rate.csv')
df
```
