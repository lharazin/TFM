# OECD Indicators - Bussiness

Exploring all indicators for Bussiness category from OECD. Described in chapter 3.2.

```python
import os
os.chdir(os.path.expanduser('../AWS_Lambdas/src/lambda_oecd_download/'))
from oecd_api import get_oecd_data
```

```python
indicator = 'MEI_CLI' # Composite Leading Indicators (MEI) 
subject = 'BSCICP03' # OECD Standardised Bussiness Confidence Indicator, Amplitude adjusted (Long term average=100), sa
measure = ''
freq = 'M'
df = get_oecd_data(indicator, subject, measure, freq, with_measure=False, 
                   file_name=f'../../../OECD_Stats/Indicators/OECD_Bussiness Confidence Indicator.csv')
df
```
