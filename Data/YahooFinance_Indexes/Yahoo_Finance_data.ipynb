```python
msci_world_country_to_index = dict()
msci_world_country_to_index.update({'United States': '^GSPC'}) # S&P 500
msci_world_country_to_index.update({'Japan': '^N225'}) # Nikkei 225
msci_world_country_to_index.update({'United Kingdom': '^FTSE'}) # FTSE 100
msci_world_country_to_index.update({'Canada': '^GSPTSE'}) # S&P/TSX Composite index
msci_world_country_to_index.update({'France': '^FCHI'}) # CAC 40
msci_world_country_to_index.update({'Switzerland': '^SSMI'}) # SMI PR
msci_world_country_to_index.update({'Germany': '^GDAXI'}) # DAX PERFORMANCE-INDEX
msci_world_country_to_index.update({'Australia': '^AXJO'}) # S&P/ASX 200
msci_world_country_to_index.update({'Netherlands': '^AEX'}) # AEX-Index
msci_world_country_to_index.update({'Denmark': '^OMXC25'}) # OMX Copenhagen 25 Index (from 2016-12-19) csv from 2001
msci_world_country_to_index.update({'Sweden': '^OMX'}) # OMX Stockholm 30 Index (from 2008-11-20) csv from 1999
msci_world_country_to_index.update({'Spain': '^IBEX'}) # IBEX 35
msci_world_country_to_index.update({'Hong Kong': '^HSI'}) # HANG SENG INDEX
msci_world_country_to_index.update({'Italy': 'FTSEMIB.MI'}) #FTSE MIB Index
msci_world_country_to_index.update({'Singapore': '^STI'}) # STI Index
msci_world_country_to_index.update({'Finland': '^OMXH25'}) # OMX Helsinki 25 (from 2013-03-05) csv from 2001
msci_world_country_to_index.update({'Belgium': '^BFX'}) # BEL 20
msci_world_country_to_index.update({'Norway': '^OSEAX'}) # Oslo Bors All-Share Index_GI (from 2013-03-05) csv from 1999
msci_world_country_to_index.update({'Israel': '^TA125.TA'}) # TA-125
msci_world_country_to_index.update({'Ireland': '^ISEQ'}) # ISEQ All Share
msci_world_country_to_index.update({'New Zealand': '^NZ50'}) # S&P/NZX 50 INDEX GROSS (from 2003-01-03) csv from 2001
msci_world_country_to_index.update({'Austria': '^ATX'}) # Austrian Traded Index
msci_world_country_to_index.update({'Portugal': 'PSI20.LS'}) # PSI 20 (from 2013-04-30) csv from 2010
```

```python
df = pd.DataFrame()

for country in msci_world_country_to_index:
    index_ticker = msci_world_country_to_index[country]
    index_data = yf.download(index_ticker, '1999-1-1')
    print(country, index_data.shape)
    #index_data.to_csv('./Indexes_1999_2023-07-28/' + country + '.csv')
    df[country] = index_data['Adj Close']
```

```python
df.to_csv('MSCI_World_Countries_Indexes.csv')
df
```

```python
for country in msci_world_country_to_index:
    first_valid = df[country].first_valid_index()
    print(country, first_valid)
```

```python
df.plot(figsize=(20,10))
```
