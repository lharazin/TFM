# Economic Calendar Backup

Recovery backup for economic calendar that involved a lot of semi-manual web scrapping. Prepared in case of the issues with AWS database.

```python
import pandas as pd
import numpy as np

from dotenv import load_dotenv
load_dotenv()
```

```python
from SqlAlquemySelectHandler import SqlAlquemySelectHandler
sql_handler = SqlAlquemySelectHandler()
```

```python
df = pd.DataFrame()

for year in range(1999, 2024):
    df_year = sql_handler.read_all_indicators(year)
    df = pd.concat([df, df_year])
df.index = np.arange(0, len(df))
df = df.iloc[:,1:]
df
```

```python
df.to_csv('investing_economic_calendar_2024-01-05.csv')
```
