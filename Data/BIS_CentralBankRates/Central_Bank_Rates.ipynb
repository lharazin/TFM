BIS (Bank of International Settlement) - Central bank policy rates

https://www.bis.org/statistics/cbpol.htm

```python
bis = pd.read_csv('BIS_CentralBankRates.csv', parse_dates=True, index_col=0)
bis.columns = bis.loc['Reference area', :]
bis.columns.name = None
bis = bis.iloc[8:,:]
bis.index.name = None
bis.index = pd.to_datetime(bis.index)
bis = bis.loc['1999':]
```

```python
bis.columns
```

```python
bis.head()
```

Singapore Daily Interest Rates

https://eservices.mas.gov.sg/Statistics/dir/DomesticInterestRates.aspx

```python
overnight = pd.read_csv('Singapore_overnight_rates_1999-2005.csv')
overnight = overnight.ffill()
overnight['Day'] = overnight['Day'].astype('int').astype('string')
overnight['Date'] = overnight['Year'].astype('int').astype('string') + '-' + overnight['Month'] + '-' + overnight['Day']
overnight.index = pd.to_datetime(overnight['Date'])
overnight['Interbank Overnight']
```

```python
sora = pd.read_csv('Singapore_SORA_2005-2023.csv', parse_dates=True, index_col=3)
sora['SORA']
```

```python
singapore_rates = pd.concat([overnight['Interbank Overnight'], sora['SORA']])
singapore_rates
```

```python
for c in msci_world_central_bank_rates.columns:
    first = msci_world_central_bank_rates[c].first_valid_index()
    print(c, first)
```

```python
msci_world_central_bank_rates = msci_world_central_bank_rates.ffill()
msci_world_central_bank_rates = msci_world_central_bank_rates.bfill()
msci_world_central_bank_rates
```

```python
msci_world_central_bank_rates.to_csv('msci_world_central_bank_rates_1999_2023-07-24.csv')
```
