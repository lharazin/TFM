```python
def split_into_chunks(df, chunk_size=200):
    chunks = list()
    num_chunks = len(df)//chunk_size + 1
    for i in range(num_chunks):
        chunks.append(df[i*chunk_size:(i+1)*chunk_size])
    return chunks
```

```python
sql_handler = SqlAlquemyInsertMarketDataHandler()
for country in central_bank_rates:
    symbol_code = sql_handler.get_symbol_code('Central Bank Rate', country)

    rates_diff = central_bank_rates[country].diff()
    rates_change = central_bank_rates[country][rates_diff != 0]
    rates_change = rates_change.dropna()
    if (len(rates_change) < 200):
        sql_handler.insert_into_table(symbol_code, rates_change)
    else:
        chunks = split_into_chunks(rates_change)
        for chunk in chunks:
            sql_handler.insert_into_table(symbol_code, chunk)

    final_count = sql_handler.get_indicator_count(symbol_code)
    print(f'{country}: Inserted', final_count, 'records')
```
