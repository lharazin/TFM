```python
import pandas as pd
```

```python
all_countries = ['United States', 'Japan', 'United Kingdom', 'Canada', 'France',
                 'Switzerland', 'Germany', 'Australia', 'Netherlands', 'Denmark',
                 'Sweden', 'Spain', 'Hong Kong', 'Italy', 'Singapore',
                 'Finland', 'Belgium', 'Norway', 'Israel', 'Ireland',
                 'New Zealand', 'Austria', 'Portugal', #'Eurozone',
                 'China', 'Taiwan', 'India', 'Korea', 'Brazil',
                 'Saudi Arabia', 'South Africa', 'Mexico', 'Thailand', 'Indonesia',
                 'Malaysia', 'United Arab Emirates', 'Qatar', 'Kuwait', 'Turkiye',
                 'Philippines', 'Poland', 'Chile', 'Greece', 'Peru',
                 'Hungary', 'Czechia', 'Egypt', 'Colombia', 'Argentina', 'Russia']
```

## Countries by Stock Market Capitalization

```python
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_countries_by_stock_market_capitalization')
total_market_cap_df = df[0]
total_market_cap_df
```

```python

total_market_cap_df.index = total_market_cap_df['Country']
total_market_cap_df.index.name = None
total_market_cap = total_market_cap_df['Total market cap (in mil. US$)[2]']
total_market_cap.name = 'Total market cap'

total_market_cap.rename(index={
    'Czech Republic': 'Czechia',
    'Turkey': 'Turkiye',
    'South Korea': 'Korea'
}, inplace=True)

total_market_cap
```

```python
# Correct annotations
for country in total_market_cap.index:
    if str(total_market_cap[country]).endswith(']'):
        total_market_cap[country] = int(total_market_cap[country][:-3].replace(',', ''))
```

```python
total_market_cap = total_market_cap.astype('int32')
total_market_cap.loc[all_countries]
```

```python
total_value = total_market_cap.sum()
total_value
```

```python
pct_of_global_stock = total_market_cap.loc[all_countries]/total_value * 100
pct_of_global_stock = round(pct_of_global_stock, 2)
pct_of_global_stock
```

```python
pct_of_global_stock.to_csv('Indicators/countries_by_stock_market_capitalization.csv')
```
